## Week 6 - Final Design
This week’s project is all about logic. We need to understand how to set up a step-by-step process for logging users in and out. There are three main parts to this program, registering a user, logging a user in, and the main loop that will run the program. Knowing that the first two are tasks, we can make functions out of them and call them when necessary in the main loop. Let’s go ahead and lay out the logical process for this program:
1.	Check to see if user is logged in.

        a. If logged in, ask if they would like to log out/quit.
   
             i.	Either quit or log out user and restart.

        b. Else, ask if they would like to log in/register/quit.
                
             i. If log in, ask user for e-mail/password.

                    1. If correct, log user in and restart.
                    
                    2. Else, display error and restart.

            ii. If register, ask for e-mail/password/password2.
                        
                    1. If passwords match, save user and restart.
                    
                    2. Else, display error and restart.
 
            iii. If quit, say thank you and exit program.

 
 
 

```python
import csv

Database={}

# the database can represent a dictionary with the following format:
# {"email"} : { ["password", log-in ]}, where log-in is boolean True for logged-in and False not logged-out



# Step 1: importing Database from CSV

def OpenCSV(): # Read from CSV
    
    global Database # IMPORTANT : variable within function must be global!!!
    
    Database={}
    
    with open("week6.csv", mode="r") as f:
    
        reader = csv.reader(f, delimiter=",")
    
        for row in reader:
        
            if len(row) == 3: # This helps to eliminate the last empty line in CSV after SaveCSV(). No other solution found :-(
                
                Database[row[0]]=[row[1],row[2]]
    
    print('Database loaded')

def SaveCSV(): # Write to CSV
    
    with open("week6.csv", mode="w") as f:
        
        writer = csv.writer(f, delimiter=",")
        
        for k, v in Database.items():
        
            writer.writerow( [k, v[0], v[1]] )
    
    print("Database saved")
            
Email = ''

while Email != "exit":            
            
    Email=input("Please input your e-mail or type 'exit' to exit: ")
    
    OpenCSV()

    if Database.get(Email,False) != False: # email in Database
    
        if Database[Email][1]=="True": # email logged in
        
            Reply=input("Do you want to logout? Y/N: ")
        
            if Reply == "Y":
                
                Database[Email][1]="False"
                
                print ("You are logged out, thank you.")
                
                SaveCSV ()
                
                continue
                
            else:
                
                continue
                
        else: # email is not logged in
            
            CheckPassword=input("Please enter your password :")
            
            if Database[Email][0]==CheckPassword:
                
                print ("Your are now logged in, thank you!")
                
                Database[Email][1]="True"
                
                SaveCSV ()
                
                continue
                
            else:
                
                print ("Error")
                
                continue
                
    else: # email not in Database
        
        Reply=input("Do you want to register? Y/N: ")
            
        if Reply=="Y":
                
            NewEmail, NewPass1, NewPass2 = input("Please enter your email, password and repeat password: ").split()
                
            if NewPass1 == NewPass2:
                    
                Database[NewEmail]=[NewPass1, "False"]
                SaveCSV ()
                continue
                    
            else:
                    
                print ("Error, passwords do not match")
                continue
            
        else:
            
            continue
```
